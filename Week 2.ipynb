```python
import pandas as pd

# Load the dataset
file_path = 'iris.csv'  # Update this path if needed
iris_data = pd.read_csv(file_path)

# 1. Explore the data
# Display the first few rows
print("First few rows of the dataset:")
print(iris_data.head())

# Check data types and column information
print("\nDataset Info:")
iris_data.info()

# Get summary statistics for numerical columns
print("\nSummary Statistics:")
print(iris_data.describe())

# 2. Calculate basic statistics
# Mean
mean_values = iris_data.mean(numeric_only=True)
print("\nMean Values:")
print(mean_values)

# Median
median_values = iris_data.median(numeric_only=True)
print("\nMedian Values:")
print(median_values)

# Mode
mode_values = iris_data.mode().iloc[0]  # Get the first mode for each column
print("\nMode Values:")
print(mode_values)

# Standard Deviation
std_values = iris_data.std(numeric_only=True)
print("\nStandard Deviation Values:")
print(std_values)

# Correlation matrix (exclude non-numerical columns)
correlation_matrix = iris_data.select_dtypes(include='number').corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
Observations:

    Petal Length and Petal Width have a strong positive correlation (0.96).
    Sepal Length correlates strongly with Petal Length (0.87) and Petal Width (0.82).
    Sepal Width has weak or negative correlations with other features.
```
